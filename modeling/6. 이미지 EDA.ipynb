```python
import numpy as np
import matplotlib.pyplot as plt
import os

# .bin  
file_dir = "C:/Users/82103/Desktop/multimodal/train(01~14)val(15~16)test(17~18)/AGV/train/01/"
file_names = [
    "agv01_0901_081240.bin",
    "agv01_0902_201540.bin",
    "agv01_0902_201554.bin",
    "agv01_0902_201608.bin",
    "agv01_0902_201627.bin",
    "agv01_0902_201651.bin",
    "agv01_0902_201718.bin",
    "agv01_0902_201741.bin",
    "agv01_0902_201759.bin",
    "agv01_0902_201817.bin"
]

#    
file_paths = [os.path.join(file_dir, fname) for fname in file_names]

#     
for i, file_path in enumerate(file_paths):
    try:
        #   
        bin_data = np.load(file_path, allow_pickle=True).astype(np.float32)
        print(f"File {i+1}: {file_path} loaded successfully. Data shape: {bin_data.shape}")

        #    (120x160)
        reshaped_data = bin_data.reshape((120, 160))

        #  
        plt.figure(figsize=(5, 5))
        plt.imshow(reshaped_data, cmap="hot")
        plt.colorbar()
        plt.title("agv01_0901_081240")
        plt.show()
    except Exception as e:
        print(f"Error loading or visualizing file {file_path}: {e}")
```
